```python
import pandas as pd

nse_data=pd.read_csv('DIRTY_DATA_NSE.csv')
nse_data.head()
```

```python
nse_data.shape
```

```python
nse_data.describe()
```

```python
nse_data.isnull().sum()
```

```python
stock_data=nse_data.dropna(subset=['SERIES'])
stock_data.isnull().sum()
```

```python
stock_data["OPEN"].fillna(stock_data["OPEN"].median(),inplace=True)
stock_data["HIGH"].fillna(stock_data["HIGH"].median(),inplace=True)
stock_data["LOW"].fillna(stock_data["LOW"].median(),inplace=True)
stock_data["CLOSE"].fillna(stock_data["CLOSE"].median(),inplace=True)
stock_data.head()
```

```python
stock_data.isnull().sum()
```

```python
from collections import OrderedDict

stock_data['TIMESTAMP'] = pd.to_datetime(stock_data['TIMESTAMP'])
stock_data = stock_data.set_index('TIMESTAMP')
stock_data = stock_data.resample('W').agg(OrderedDict([
        ('OPEN', 'first'),
        ('HIGH', 'max'),
        ('LOW', 'min'),
        ('CLOSE', 'last'),
    ])
)

stock_data.head()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(stock_data['OPEN'])
plt.plot(stock_data['HIGH'])
plt.plot(stock_data['LOW'])
plt.plot(stock_data['CLOSE'])
plt.show()
```
