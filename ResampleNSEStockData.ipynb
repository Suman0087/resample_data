```python
from pyspark.sql.functions import window
from pyspark.sql import SparkSession
import pyspark.sql.functions as F
from pyspark.sql.functions import col

spark = SparkSession.builder.getOrCreate()

nsestock_data = spark.read.csv('FINAL_FROM_DF.csv',inferSchema=True,header=True)

df_data = nsestock_data \
            .groupBy(F.window("TIMESTAMP", "1 week")) \
            .agg(
                F.first('OPEN').alias('OPEN'),
                F.max('HIGH').alias('HIGH'),
                F.min('LOW').alias('LOW'),
                F.last('CLOSE').alias('CLOSE'),
                )

df=df_data.orderBy(col("window")).show(truncate=False)
```
